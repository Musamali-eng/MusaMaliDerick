```python
#Absolute Maximum Scalling
import pandas as pd
import numpy as np

df = pd.read_csv(r'D:\Recess\MusaMaliDerick\Lectures\week4\Morning_DataScience\Housing.csv')
df = df.select_dtypes(include=[np.number])

print(df.head())
```

```python
#selecting only numeric columns
max_abs = np.max(np.abs(df),axis =0)
print(max_abs)

#Divides each value by the maximum absolute value of its column to scale the data.
scaled_df = df / max_abs
print(scaled_df)
```
