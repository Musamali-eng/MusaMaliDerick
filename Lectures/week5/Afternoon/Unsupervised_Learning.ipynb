```python
import numpy as np
from sklearn.cluster import KMeans

```

```python
#unsupervised learning is a type of machine learning were the model is trained on unlabed data
#the goal is to find hidden patterns or intrinsic strutures in the input data
#unlike supervised learning  where the model is trained  on label data, unsupervised learning does not  have predefined output labels.
#Realworld application:Grouping customers based on purchasing behavior.
#Anomaly Detection: identifying unsual patterns in a dataset while preserving important info , often used for data visualization
#Recommended system: suggesting products or content to users based on their preference and behavior
#image and video Aanalysis: Grouping similar images or  videos together for organization of retrieval.
#Document Clustering: Organizing large collection of documents into meaningful clusters for easier navigation and search
#Medical Diagnosis: Identifying patterns in medical data to assist in diagnosis and treatment planning.
#Social Network Analyis: Understanding relationships and communities within social networks.
#Market Basket Ananalysis: Discovering associations btn products purchased togather by customers.
#Generic data analysis : Identifying patterns in genericdata for research and personalized medicine.
#Network security:Detecting unusal network traffic patterns that may indicate security pattern.
```

```python
#Example Image Clustering using K-Means (unsupervised learning)
#What is Clustering 
#Clustering is an unsupervised machine learning technique that groups unlabeled data into clusters based on similarity. Its goal is to discover patterns or relationships within the data without any prior knowledge of categories or labels.

#Groups data points that share similar features or characteristics.
#Helps find natural groupings in raw, unclassified data.
#Commonly used for customer segmentation, anomaly detection and data organization.
#Works purely from the input data without any output labels.
#Enables understanding of data structure for further analysis or decision-making.
#What is K-Means
image_names = [
    #image names is only used for displaying result

"dog_1.png","dog_2.png","dog_3.png",

"dog_4.png","dog_5.png","dog_6.png",

"cat_1.png","cat_2.png","cat_3.png",

"cat_4.png","cat_5.png","cat_6.png",

"car_1.png","car_2.png","car_3.png",

"car_4.png","car_5.png","car_6.png",

"mountain_1.png","mountain_2.png","mountain_3.png",

"mountain_4.png","mountain_5.png","mountain_6.png",

"unknown_1.png",

"unknown_2.png",

"unknown_3.png",

"unknown_4.png",

"unknown_5.png",

"unknown_6.png"

]

```

```python
image_features = np.array([

# Dogs

[12,18],[14,20],[13,17],[15,19],[11,18],[16,20],

# Cats

[60,65],[62,67],[59,64],[63,66],[61,63],[60,68],

# Cars

[120,20],[118,18],[122,22],[119,19],[121,21],[123,20],

# Mountains

[200,180],[198,182],[201,179],

[203,181],[199,178],[202,183],

# Unknown Images

[250,50], # Flower

[180,20], # Airplane

[170,100], # Bicycle

[90,150], # City

[230,90], # Burger

[180,200] # Beach

])


```

```python
kmeans = KMeans(
    n_clusters=4,
    random_state=42
)

# Train the model

kmeans.fit(image_features)

# Cluster assignments

clusters = kmeans.labels_
```

```python

print("IMAGE CLUSTERING RESULTS")

for image, cluster in zip(image_names, clusters):

    print(f"{image:18} ---> Cluster {cluster}")
    print("\nCluster Centers")

print(kmeans.cluster_centers_)
```
